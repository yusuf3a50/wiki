# Creating a Simple Decorator
To create a decorator, you define a function that takes another function as an argument, defines a nested wrapper function, and returns this wrapper. Hereâ€™s a basic example:

```python
def simple_decorator(func):
    def wrapper():
        print("Before the function call")
        func()
        print("After the function call")
    return wrapper
```

You can apply this decorator to a function using the @ syntax:

```python
@simple_decorator
def greet():
    print("Hello, World!")
```

When you call greet(), it will execute the wrapper, adding behavior before and after the original function.
